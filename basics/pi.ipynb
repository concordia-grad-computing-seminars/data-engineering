# Computing the number $\pi$ using Monte Carlo simulation

## Basic idea

![Darts](https://raw.githubusercontent.com/coderefinery/jupyter/main/example/darts.svg)

We can evaluate the value of $\pi$ if we compute the ratio of the surface of a unit circle and a unit square.

Relevant formulas
- Square area: $S = (2 r)^2$
- Circle area: $C = \pi r^2$
- $C/S = (\pi r^2) / (4 r^2) = \pi / 4$
- $\pi = 4 \cdot C/S$

## Libraries

```python
import random
```

## Code

```python
# initializing the number of "throws"
num_points = 10**7
```

```python
# "throw darts" and count the number of hits
hits = 0
for _ in range(num_points):
    x, y = random.random(), random.random()
    if x*x + y*y < 1.0:
        hits += 1
```

```python
# compute and print the estimate
fraction = hits / num_points
4*fraction
```

## Visualisation

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
num_points = 10**3
points = []
hits = 0
for _ in range(num_points):
    x, y = random.random(), random.random()
    if x*x + y*y < 1.0:
        hits += 1
        points.append((x, y, "red"))
    else:
        points.append((x, y, "blue"))
```

```python
# unzip points into 3 lists
x, y, colors = zip(*points)

# define figure dimensions
fig, ax = plt.subplots()
fig.set_size_inches(6.0, 6.0)

# plot results
ax.scatter(x, y, c=colors)

plt.show()
```
