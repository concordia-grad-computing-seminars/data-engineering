## Strings

### Definition

Strings are defined like this (either using `"` or `'`'):

````python
s1 = "Hello"
s2 = 'World'
````

Long strings can be split over several lines like this:<br>
(source for text: https://www.lyricsondemand.com/tvthemes/vikingslyrics.html)

````python
long_str = """
More, give me more, give me more
If I had a heart I could love you
If I had a voice I would sing
After the night when I wake up
I'll see what tomorrow brings
"""

print(long_str)
````

## Sets

### Definition

<div class="alert alert-block alert-info">

## Dictionaries

### Definition

<div class="alert alert-block alert-info">
    Python <a href="https://docs.python.org/3/tutorial/datastructures.html#dictionaries">dictionaries </a> are data structures indexed by keys.
</div>
They are defined like this:

````python
dic1 = {'oranges': 10, 
        'bananas': 15, 
        'apples': 3}
````

Alternatively one can use the `dict()` constructor. 
<br>
The following examples produce all the same dictionary. 
<br>
This is useful to know, as sometimes one has different types of data one wishes to transform into a dictionary.

````python
dict(oranges=10, bananas=15, apples=3)
````

````python
# Constructing a dictionnary from a list of tuples
dict([('oranges', 10), ('bananas', 15), ('apples', 3)])
````

### Accessing elements using keys

The key is used to get an element:

````python
dic1['bananas']
````

or like this

````python
dic1.get('bananas')
````

The key can be used to update or add new elements:

````python
dic1['apples'] = 23
dic1['pears'] = 5
print(dic1)
````

Keys are listed like this:

````python
list(dic1)
````

### Iterables for dictionaries

Dictionaries can be looped using the `items()` `iterable`:

````python
for i in dic1.items():
    print(i)
````

Or like this, if you want to access directly the keys and values

````python
for k, v in dic1.items():
    print(k, v)
````

The key `iterable` is `keys()`

````python
for k in dic1.keys():
    print(k)
````

The values `iterable` is `values()`:

````python
for v in dic1.values():
    print(v)
````

### An example: counting words in a text

The definition of data science according Wikipedia:

````python
text = '''
Data science is an interdisciplinary academic field that uses statistics, scientific computing, scientific methods, processes, algorithms and systems
to extract or extrapolate knowledge and insights from noisy, structured and unstructured data.
Data science also integrates domain knowledge from the underlying application domain (e.g., natural sciences, information technology, medicine). 
Data science is multifaceted and can be described as a science, as a research paradigm, as a research method, as a discipline, as a workflow, and as a profession.

Data science is a "concept to unify statistics, data analysis, informatics, and their related methods" in order to "understand and analyse actual phenomena" 
with data. It uses techniques and theories drawn from many fields within the context of mathematics, statistics, computer science, information science, 
and domain knowledge. However, data science is different from computer science and information science. 
Turing Award winner Jim Gray imagined data science as a "fourth paradigm" of science (empirical, theoretical, computational, and now data-driven) and asserted
that "everything about science is changing because of the impact of information technology" and the data deluge
'''
````

````python
word_counts = {}
for word in text.replace('\n',' ').split(' '):
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1
````

````python
word_counts
````
