## Libraries

```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
from sklearn.model_selection import RepeatedKFold

import matplotlib.pyplot as plt
%matplotlib inline
```

## Synthetic data set

To illustrate the concepts we will use a synthetic data set which we generate using the [`make_moons`](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_moons.html) function

```python
X, y = make_moons(200, noise=0.3, random_state=1)

colors = np.array(['tab:blue', 'tab:orange'])
plt.scatter(X[:, 0], X[:, 1], s=10, color=colors[y])
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.show()
```

## Classification tree training

Without any extra restrictions, decision trees tends to not fit adequately.

```python
clf = DecisionTreeClassifier()
clf.fit(X, y)
```

```python
def plot_dec_bound(clf, X, y, ax):
    '''
    Plot decision boundaries
    '''
    plot_step = 0.02

    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = ax.contourf(xx, yy, Z-100, cmap=plt.cm.Greens)

    colors = np.array(['tab:blue', 'tab:orange'])
    ax.scatter(X[:, 0], X[:, 1], s=12, color=colors[y])
    
    ax.set_xlabel('$x_1$', fontsize=14)
    ax.set_ylabel('$x_2$', fontsize=14)
```

```python
fig, ax = plt.subplots(1, 1)
plot_dec_bound(clf, X, y, ax)
plt.show()
```

Depending on the value of the hyper-parameter `min_samples_leaf` (minimum number of samples required to be at a leaf node), the classification can be over-fitted, optimal (more or less) or under-fitted.

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
i = 0

for n in [1, 5, 15]:
    clf = DecisionTreeClassifier(min_samples_leaf=n)
    clf.fit(X, y)
    plot_dec_bound(clf, X, y, ax[i])
    ax[i].set_title('min_samples_leaf = ' + str(n))
    i = i+1
    
plt.tight_layout()
plt.show()
```

Cross validation can be used for tuning the hyper-parameter `min_samples_leaf` (minimum number of samples required to be at a leaf node).

```python
loss_train = []
loss_valid = []

n_max = 16
n_range = range(1, n_max)
for n in n_range:
    clf = DecisionTreeClassifier(min_samples_leaf=n)
    clf.fit(X, y)
    scores = cross_validate(clf, X, y,
                            cv=RepeatedKFold(n_splits=10, n_repeats=50),
                            scoring='accuracy', return_train_score=True)
    loss_train = np.append(loss_train, scores['train_score'].mean())
    loss_valid = np.append(loss_valid, scores['test_score'].mean())

plt.plot(n_range, loss_train, "o-", markersize=8, markerfacecolor='white', label='Training set')
plt.plot(n_range, loss_valid, "s-", markersize=8, markerfacecolor='white', label='Validation set')
plt.xlabel('min_samples_leaf', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.legend(loc='best')
plt.show()
```

```python
n_opt = n_range[np.argmax(loss_valid)]

clf = DecisionTreeClassifier(min_samples_leaf=n_opt)
clf.fit(X, y)
```

```python
fig, ax = plt.subplots(1, 1)
plot_dec_bound(clf, X, y, ax)
plt.show()
```
