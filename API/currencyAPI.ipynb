# Currency API

## Description

This notebook illustrates how to access a simple online API for converting currencies.

The link of the API is:
https://open.er-api.com/v6/latest/CAD

This API returns [`JSON`](https://www.json.org) data

## Practical examples

```python
import requests
```

```python
# retrieve some data using a request
url = "https://open.er-api.com/v6/latest/CAD"
req = requests.get(url)
data = req.json()
```

Let us check how looks the data

```python
data
```

This will return the rates

```python
data['rates']
```

And this returns the rate of a specific currency

```python
data['rates']['USD']
```
